```python
def read_portfolio(file, portfolio, if_train):
    df = pd.read_csv(file)
    print(df.columns.tolist())
#     Categorical features selected
#     Vehicle Performance
    vp_high = df[df['Vehicle_Performance'] == 'High'].shape[0]/df.shape[0]
    vp_intermediate = df[df["Vehicle_Performance"] == "Intermediate"].shape[0]/df.shape[0]
    vp_sports = df[df["Vehicle_Performance"] == "Sports"].shape[0]/df.shape[0]
    vp_sports_p = df[df["Vehicle_Performance"] == "Sports Premium"].shape[0]/df.shape[0]
    vp_std = df[df["Vehicle_Performance"] == "Standard"].shape[0]/df.shape[0]
#     Vehicle Anti Theft Device
    vatd_na = df[df["Vehicle_Anti_Theft_Device"] == "Not Applicable"].shape[0]/df.shape[0]
    vatd_passive = df[df["Vehicle_Anti_Theft_Device"] == "Passive Disabling-Vehicle Recovery"].shape[0]/df.shape[0]
    vatd_alarm = df[df["Vehicle_Anti_Theft_Device"] == "Alarm Only"].shape[0]/df.shape[0]
    vatd_active = df[df["Vehicle_Anti_Theft_Device"] == "Active Disabling"].shape[0]/df.shape[0]
#     Vehicle Passive Restraint
    vpr_n = df[df["Vehicle_Passive_Restraint"] == "N"].shape[0]/df.shape[0]
    vpr_y = df[df["Vehicle_Passive_Restraint"] == "Y"].shape[0]/df.shape[0]
#     Vehicle Youthful Driver Indicator
    vydi_n = df[df["Vehicle_Youthful_Driver_Indicator"] == "N"].shape[0]/df.shape[0]
    vydi_y = df[df["Vehicle_Youthful_Driver_Indicator"] == "Y"].shape[0]/df.shape[0]
#     Vehicle Youthful Driver Training_Code
    vydtc_na = df[df["Vehicle_Youthful_Driver_Training_Code"] == "Not Applicable"].shape[0]/df.shape[0]
    vydtc_wdt = df[df["Vehicle_Youthful_Driver_Training_Code"] == "With Driver Training"].shape[0]/df.shape[0]
    vydtc_wwdt = df[df["Vehicle_Youthful_Driver_Training_Code"] == "With or Without Driver Training"].shape[0]/df.shape[0]
    vydtc_wodt = df[df["Vehicle_Youthful_Driver_Training_Code"] == "Without Driver Training"].shape[0]/df.shape[0]
#     Vehicle Youthful Good Student Code
    vygsc_egsc = df[df["Vehicle_Youthful_Good_Student_Code"] == "Eligible for Good Student Credit"].shape[0]/df.shape[0]
    vygsc_negsc = df[df["Vehicle_Youthful_Good_Student_Code"] == "Not Eligible for Good Student Credit"].shape[0]/df.shape[0]
#     SYS New Business
    snb_n = df[df["SYS_New_Business"] == "N"].shape[0]/df.shape[0]
    snb_y = df[df["SYS_New_Business"] == "Y"].shape[0]/df.shape[0]
#     Vehicle Safe Driver Discount Indicator
    vsddi_n = df[df["Vehicle_Safe_Driver_Discount_Indicator"] == "N"].shape[0]/df.shape[0]
    vsddi_y = df[df["Vehicle_Safe_Driver_Discount_Indicator"] == "Y"].shape[0]/df.shape[0]
#     SYS Renewed
    sr_n = df[df["SYS_Renewed"] == "N"].shape[0]/df.shape[0]
    sr_y = df[df["SYS_Renewed"] == "Y"].shape[0]/df.shape[0]
#     Numerical features selected
#     Driver Total
    dt_mean = df["Driver_Total"].mean()
    dt_std = df["Driver_Total"].std()
#     Driver Total Female
    dtf_mean = df["Driver_Total_Female"].mean()
    dtf_std = df["Driver_Total_Female"].std()
#     Driver_Total_Single
    dts_mean = df["Driver_Total_Single"].mean()
    dts_std = df["Driver_Total_Single"].std()
#     Driver Total Related To Insured Spouse
    dtris_mean = df["Driver_Total_Related_To_Insured_Spouse"].mean()
    dtris_std = df["Driver_Total_Related_To_Insured_Spouse"].std()
#     Driver Total Related To Insured Child
    dtric_mean = df["Driver_Total_Related_To_Insured_Child"].mean()
    dtric_std = df["Driver_Total_Related_To_Insured_Child"].std()
#     Driver Total Licensed  In State
    dtls_mean = df["Driver_Total_Licensed_In_State"].mean()
    dtls_std = df["Driver_Total_Licensed_In_State"].std()
#     Driver Total Teenager Age 15_19
    dtta1519_mean = df["Driver_Total_Teenager_Age_15_19"].mean()
    dtta1519_std = df["Driver_Total_Teenager_Age_15_19"].std()
#     Driver_Total_Senior_Ages_65_69
    dtsa6569_mean = df["Driver_Total_Senior_Ages_65_69"].mean()
    dtsa6569_std = df["Driver_Total_Senior_Ages_65_69"].std()
#     Vehicle Territory
    vt_mean = df["Vehicle_Territory"].mean()
    vt_std = df["Vehicle_Territory"].std()
#     Vehicle Number Of Drivers Assigned
    vnda_mean = df["Vehicle_Number_Of_Drivers_Assigned"].mean()
    vnda_std = df["Vehicle_Number_Of_Drivers_Assigned"].std()
#     Driver Total Adult Ages 50_64
    dtaa5064_mean = df["Driver_Total_Adult_Ages_50_64"].mean()
    dtaa5064_std = df["Driver_Total_Adult_Ages_50_64"].std()
#     EEA Policy Tenure
    ept_mean = df["EEA_Policy_Tenure"].mean()
    ept_std = df["EEA_Policy_Tenure"].std()
#     Annual Premium
    ap_mean = df["Annual_Premium"].mean()
    ap_std = df["Annual_Premium"].std()
    if if_train:
#     Loss Amount
        la_mean = df["Loss_Amount"].mean()
        la_std = df["Loss_Amount"].std()
    if not if_train:
        summary = {"ID": str(portfolio), 
                   "vp_high": vp_high, 
                   "vp_intermediate": vp_intermediate, 
                   "vp_sports": vp_sports, 
                   "vp_sports_p": vp_sports_p,  
                   "vp_std": vp_std, 
                   "vatd_na": vatd_na, 
                   "vatd_passive": vatd_passive, 
                   "vatd_alarm": vatd_alarm, 
                   "vatd_active": vatd_active, 
                   "vpr_n": vpr_n,
                   "vpr_y": vpr_y,
                   "vydi_n": vydi_n, 
                   "vydi_y": vydi_y, 
                   "vydtc_na": vydtc_na, 
                   "vydtc_wdt": vydtc_wdt,  
                   "vydtc_wwdt": vydtc_wwdt, 
                   "vydtc_wodt": vydtc_wodt, 
                   "vygsc_egsc": vygsc_egsc, 
                   "vygsc_negsc": vygsc_negsc, 
                   "snb_n": snb_n, 
                   "snb_y": snb_y,
                   "vsddi_n": vsddi_n,
                   "vsddi_y": vsddi_y, 
                   "sr_n": sr_n, 
                   "sr_y": sr_y, 
                   "dt_mean": dt_mean,  
                   "dt_std": dt_std, 
                   "dtf_mean": dtf_mean, 
                   "dtf_std": dtf_std, 
                   "dts_mean": dts_mean,  
                   "dts_std": dts_std, 
                   "dtris_mean": dtris_mean, 
                   "dtris_std": dtris_std,  
                   "dtric_mean": dtric_mean, 
                   "dtric_std": dtric_std,
                   "dtls_mean": dtls_mean,  
                   "dtls_std": dtls_std, 
                   "dtta1519_mean": dtta1519_mean, 
                   "dtta1519_std": dtta1519_std, 
                   "dtsa6569_mean": dtsa6569_mean,  
                   "dtsa6569_std": dtsa6569_std, 
                   "vt_mean":vt_mean, 
                   "vt_std": vt_std,  
                   "vnda_mean": vnda_mean, 
                   "vnda_std": vnda_std,
                   "dtaa5064_mean": dtaa5064_mean, 
                   "dtaa5064_std": dtaa5064_std, 
                   "ept_mean": ept_mean, 
                   "ept_std": ept_std,
                   "ap_mean": ap_mean, 
                   "ap_std": ap_std}
    else:
        summary = {"ID": str(portfolio), 
                   "vp_high": vp_high, 
                   "vp_intermediate": vp_intermediate, 
                   "vp_sports": vp_sports, 
                   "vp_sports_p": vp_sports_p,  
                   "vp_std": vp_std, 
                   "vatd_na": vatd_na, 
                   "vatd_passive": vatd_passive, 
                   "vatd_alarm": vatd_alarm, 
                   "vatd_active": vatd_active, 
                   "vpr_n": vpr_n,
                   "vpr_y": vpr_y,
                   "vydi_n": vydi_n, 
                   "vydi_y": vydi_y, 
                   "vydtc_na": vydtc_na, 
                   "vydtc_wdt": vydtc_wdt,  
                   "vydtc_wwdt": vydtc_wwdt, 
                   "vydtc_wodt": vydtc_wodt, 
                   "vygsc_egsc": vygsc_egsc, 
                   "vygsc_negsc": vygsc_negsc, 
                   "snb_n": snb_n, 
                   "snb_y": snb_y,
                   "vsddi_n": vsddi_n,
                   "vsddi_y": vsddi_y, 
                   "sr_n": sr_n, 
                   "sr_y": sr_y, 
                   "dt_mean": dt_mean,  
                   "dt_std": dt_std, 
                   "dtf_mean": dtf_mean, 
                   "dtf_std": dtf_std, 
                   "dts_mean": dts_mean,  
                   "dts_std": dts_std, 
                   "dtris_mean": dtris_mean, 
                   "dtris_std": dtris_std,  
                   "dtric_mean": dtric_mean, 
                   "dtric_std": dtric_std,
                   "dtls_mean": dtls_mean,  
                   "dtls_std": dtls_std, 
                   "dtta1519_mean": dtta1519_mean, 
                   "dtta1519_std": dtta1519_std, 
                   "dtsa6569_mean": dtsa6569_mean,  
                   "dtsa6569_std": dtsa6569_std, 
                   "vt_mean":vt_mean, 
                   "vt_std": vt_std,  
                   "vnda_mean": vnda_mean, 
                   "vnda_std": vnda_std,
                   "dtaa5064_mean": dtaa5064_mean, 
                   "dtaa5064_std": dtaa5064_std, 
                   "ept_mean": ept_mean, 
                   "ept_std": ept_std,
                   "ap_mean": ap_mean, 
                   "ap_std": ap_std}
        summary["la_mean"]= la_mean
        summary["la_std"]=la_std
    return summary
```
